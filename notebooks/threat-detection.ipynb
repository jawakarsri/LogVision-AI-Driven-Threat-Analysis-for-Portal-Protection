# Binary Classification using Anomaly Detection

First, we will use Isolation Forest algorithm for anomaly detection on a scaled version of the dataset, in order to identify anomalies or malicious logs.

```python
X=df.drop('threat',axis=1)
y=df['threat'].apply(lambda x: 0 if x==0 else 1)
X
```

We will use MinMaxScaler to scale the dataset.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pickle

pickle.dump(MinMaxScaler().fit(X), open('scaler.pkl', 'wb'))
X = pd.DataFrame(MinMaxScaler().fit_transform(X))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train
```

We will use Isolation Forest algorithm for anomaly detection. We will use n_estimators=800, bootstrap=True as hyperparameters.

```python
from sklearn.ensemble import IsolationForest as iso

model = iso(n_estimators=800, bootstrap=True, random_state=69)
model.fit(X_train)

y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = pd.Series(y_pred)
y_pred = y_pred.apply(lambda x: 1 if x == -1 else 0)
accuracy_score(y_test, y_pred)
```

We get an accuracy of 92.2% which is quite good given that we have a huge and imbalanced dataset. However, in order to further analyze the results, we will use classification report and confusion matrix.

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

We can see that the model is not under-delivering, in the sense that it is not letting any genuine threats pass through. However we can see an error rate of 8.8% which is okay, given the size and quality of the dataset. Hence it is desirable that the model produce this error in overdetecting the threats rather than underdetecting them.
